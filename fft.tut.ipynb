```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_t = 100
x_t = np.zeros(num_t)
x_t[:10] = 1

plt.plot(x_t)
plt.title("0-9 Pulse function")
plt.xlabel("Sample Idx")
plt.ylabel("Amplitude(Val) of x")
```

```python
f_x_t = np.fft.fft(x_t)
plt.subplot(1,2,1) #real part of fourier transform plot
plt.plot(np.fft.fftshift(np.real(f_x_t)))
plt.title("Real Part of Fourier Transform of x_t")
plt.ylabel("Value/Magnitude")
plt.xlabel("Frequency Idx")

plt.subplot(1,2,2) #imaginary part of fourier transform plot
plt.plot(np.fft.fftshift(np.imag(f_x_t)))
plt.title("Imaginary Part of Fourier Transform of x_t")
plt.ylabel("Value/Magnitude")
plt.xlabel("Frequency Idx")
```
